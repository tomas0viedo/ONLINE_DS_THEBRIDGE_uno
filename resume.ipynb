# Resumen del Proyecto: Predicción de Demanda de Productos

Este cuaderno presenta un resumen ejecutivo de un proyecto de Machine Learning centrado en la predicción de la demanda diaria de productos. Exploraremos el objetivo, los datos, los modelos utilizados, los resultados clave y las variables más influyentes.

## Objetivo y Datos

**Objetivo del Modelo Predictivo:** Estimar la cantidad de productos vendidos por departamento cada día para optimizar la planificación operativa y la toma de decisiones estratégicas.

**Contexto y Origen de Datos:** El proyecto se basa en el **DataCoSupplyChainDataset**, un extenso conjunto de datos de cadena de suministro que abarca ventas, pedidos y beneficios desde 2015 hasta 2018.

## Procesamiento de Datos y Feature Engineering

Se implementó un robusto proceso de preprocesamiento de datos, que incluyó:
* Limpieza de datos (manejo de nulos y duplicados).
* Generación de **variables temporales** (año, mes, día de la semana) para capturar patrones estacionales y diarios.
* **Codificación One-Hot** para variables categóricas como el nombre del departamento.
* **Escalado numérico** de las características para mejorar el rendimiento del modelo.

## Modelos Evaluados y Métricas Clave

Se evaluaron tres modelos de regresión supervisada, dividiendo los datos en un 80% para entrenamiento y un 20% para prueba:
* **Random Forest:** Un modelo de ensamblaje robusto.
* **XGBoost:** Conocido por su alta precisión y eficiencia.
* **LightGBM:** Optimizado para grandes datasets.

Las **Métricas de Evaluación** utilizadas fueron:
* **R² (Coeficiente de Determinación):** Mide el poder explicativo del modelo.
* **RMSE (Error Cuadrático Medio):** Cuantifica la magnitud promedio de los errores de predicción.
* **MAE (Error Absoluto Medio):** Indica la desviación promedio absoluta entre las predicciones y los valores reales.

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

image_path = 'comparacion_metricas_modelos_final.png'
img = mpimg.imread(image_path)

plt.figure(figsize=(12, 6)) 
plt.imshow(img)
plt.axis('off') 
plt.show()
```

## Resultados y Conclusiones Clave

Los resultados demuestran la efectividad de los modelos en la captura de tendencias de demanda.
* **XGBoost:** Alcanzó el mejor R² (0.85), indicando un alto poder explicativo y una buena capacidad para capturar la variabilidad de la demanda.
* **Random Forest:** Mostró el menor RMSE (4.63) y MAE (2.07), lo que significa que, en promedio, sus predicciones son las más cercanas a los valores reales.

Esto sugiere que, aunque XGBoost explica mejor la varianza, Random Forest podría ser ligeramente superior en términos de la magnitud del error promedio, siendo una excelente opción para la predicción de la demanda diaria.

## Variables Más Relevantes

Hemos identificado las variables que tienen la mayor influencia en la predicción de la demanda:
* **Ventas Totales:** El factor más influyente, directamente correlacionado con la demanda futura.
* **Ganancia por Orden:** Impacta directamente la cantidad de productos demandados, sugiriendo que la rentabilidad por artículo es un predictor clave.
* **Departamento:** La categoría del departamento influye significativamente en la demanda, mostrando variaciones según la sección de productos.
* **Variables Temporales (Día de la semana y Mes):** Revelan patrones de demanda estacionales y diarios, cruciales para entender las fluctuaciones a lo largo del tiempo.
